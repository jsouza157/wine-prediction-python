```python
import pandas as pd
dataset = pd.read_csv("wine.csv")
```

```python
dataset.head()
```

```python
dataset['style'] = dataset['style'].replace('red', 0)
```

```python
dataset['style'] = dataset['style'].replace('white', 1)
```

```python
dataset.head()
```

```python
#Y é meu target e variavel X é meu preditor
y = dataset['style']
x = dataset.drop('style', axis = 1)
```

```python
#Utilizando o modelo de seleção para desocbir qual vinho é tinto e qual não é
from sklearn.model_selection import train_test_split

#Criando meu conjunto de treino e teste
x_training, x_test, y_training, y_test = train_test_split(x, y, test_size = 0.3)
#nessa parte a própria biblioteca faz o treino com os dados.
```

```python
#ExtraTreesClassifier = árvore de decisão
from sklearn.ensemble import ExtraTreesClassifier

#criando meu modelo
model = ExtraTreesClassifier()
#aplicando a árvore de decisão nos meus dados de treino com o comando fit
model.fit(x_training, y_training)

#Após o treino vou exibir os dados para garantir que a classificação funcinou funcionou
result = model.score(x_test, y_test)
print(result)
```
